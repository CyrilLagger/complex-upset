Jupyter-specific: set plot sizes:

```r
options(repr.plot.width=10, repr.plot.height=5, repr.plot.res=150, jupyter.plot_mimetypes="image/svg+xml")
```

Get the movies data:

```r
if(!require(ggplot2movies)) install.packages('ggplot2movies')
```

```r
movies = ggplot2movies::movies
head(movies)
```

Specify which columns define the class membership:

```r
genres = c('Action', 'Animation', 'Comedy', 'Drama', 'Documentary', 'Romance', 'Short')
```

```r
library(ComplexUpset)
library(ggplot2)
```

```r
upset(
    movies, genres,
    base_annotations=list('Intersection size'=intersection_size(counts=FALSE)),
    min_size=100,
    width_ratio=0.1
)
```

To remove the warning about conversion use:

```r
movies[, genres] = sapply(movies[, genres], as.logical)
```
